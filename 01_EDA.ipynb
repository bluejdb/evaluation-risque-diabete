```python
# 1. Import des librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
# 2. Aperçu rapide
print("Dimensions du dataset :", df.shape)
print("\nAperçu des premières lignes :")
print(df.head())
```

```python
# 3. Informations générales
print("\nInfos sur le dataset :")
print(df.info())

print("\nStatistiques descriptives :")
print(df.describe())
```

```python
# 4. Vérification des valeurs nulles
print("\nValeurs manquantes :")
print(df.isnull().sum())
```

```python
# 5. Distribution des classes (Outcome)
print("\nDistribution de la variable cible (Outcome) :")
print(df['Outcome'].value_counts())
```

```python
sns.set(style="whitegrid")

sns.countplot(x='Outcome', data=df)
plt.title("Répartition des classes (0 = Non diabétique, 1 = Diabétique)")
plt.show()

df.hist(bins=20, figsize=(15, 10), color='skyblue')
plt.suptitle("Distribution des variables")
plt.show()

plt.figure(figsize=(15, 10))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df, x='Outcome', y=column)
    plt.title(f'{column} vs Outcome')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```
