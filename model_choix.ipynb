```python
# 1. Import des librairies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, learning_curve
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
import xgboost as xgb
import lightgbm as lgb

# 2. Chargement du dataset
df = pd.read_csv("diabetes_cleaned_v2.csv")

# 3. SÃ©paration des features (X) et de la cible (y)
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# 4. Normalisation des donnÃ©es
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. SÃ©paration en donnÃ©es d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 6. Appliquer SMOTE pour Ã©quilibrer les classes
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 7. Initialisation des modÃ¨les de base
models = {
    'SVM': SVC(kernel='linear', C=10, gamma='scale'),
    'RÃ©gression Logistique': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=200, max_depth=8, min_samples_split=5),
    'k-NN (k=5)': KNeighborsClassifier(n_neighbors=5),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    'LightGBM': lgb.LGBMClassifier()
}

# 8. EntraÃ®nement et Ã©valuation des modÃ¨les avec Cross-Validation
model_scores = {}
for model_name, model in models.items():
    # Cross-validation sur l'ensemble des donnÃ©es d'entraÃ®nement
    cv_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=5, scoring='accuracy')
    model_scores[model_name] = cv_scores.mean()

# 9. Affichage des performances des modÃ¨les
for model_name, score in model_scores.items():
    print(f"{model_name}: {score:.4f}")

# 10. Comparaison graphique des modÃ¨les
plt.figure(figsize=(10,6))
sns.barplot(x=list(model_scores.keys()), y=list(model_scores.values()))
plt.title("Performance des diffÃ©rents modÃ¨les (Accuracy)")
plt.xlabel("ModÃ¨le")
plt.ylabel("Accuracy")
plt.show()

# 11. Ãvaluation du surapprentissage (Overfitting)
train_scores = {}
test_scores = {}
for model_name, model in models.items():
    model.fit(X_train_resampled, y_train_resampled)
    train_accuracy = model.score(X_train_resampled, y_train_resampled)
    test_accuracy = model.score(X_test, y_test)
    
    train_scores[model_name] = train_accuracy
    test_scores[model_name] = test_accuracy

# 12. Affichage du surapprentissage
overfitting_data = pd.DataFrame({
    'ModÃ¨le': list(models.keys()),
    'Accuracy EntraÃ®nement': list(train_scores.values()),
    'Accuracy Test': list(test_scores.values())
})

print("\nÃvaluation du Surapprentissage (Overfitting) :")
print(overfitting_data)

plt.figure(figsize=(10,6))
sns.barplot(x='ModÃ¨le', y='Accuracy EntraÃ®nement', data=overfitting_data, color='blue', alpha=0.6, label='EntraÃ®nement')
sns.barplot(x='ModÃ¨le', y='Accuracy Test', data=overfitting_data, color='red', alpha=0.6, label='Test')
plt.title("Surapprentissage (Overfitting) des modÃ¨les")
plt.xlabel("ModÃ¨le")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# 13. Optimisation de XGBoost
param_grid_xgb = {
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
}

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
grid_search_xgb = GridSearchCV(estimator=xgb_model, param_grid=param_grid_xgb, 
                               scoring='accuracy', cv=5, verbose=2, n_jobs=-1)
grid_search_xgb.fit(X_train_resampled, y_train_resampled)
print(f"Meilleurs paramÃ¨tres pour XGBoost: {grid_search_xgb.best_params_}")
print(f"Meilleure prÃ©cision pour XGBoost: {grid_search_xgb.best_score_:.4f}")
best_xgb_model = grid_search_xgb.best_estimator_

# 14. Optimisation de LightGBM
param_grid_lgb = {
    'num_leaves': [31, 50, 100],
    'max_depth': [-1, 5, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
}

lgb_model = lgb.LGBMClassifier()
grid_search_lgb = GridSearchCV(estimator=lgb_model, param_grid=param_grid_lgb, 
                               scoring='accuracy', cv=5, verbose=2, n_jobs=-1)
grid_search_lgb.fit(X_train_resampled, y_train_resampled)
print(f"Meilleurs paramÃ¨tres pour LightGBM: {grid_search_lgb.best_params_}")
print(f"Meilleure prÃ©cision pour LightGBM: {grid_search_lgb.best_score_:.4f}")
best_lgb_model = grid_search_lgb.best_estimator_

# 15. Ãvaluation du modÃ¨le optimisÃ© XGBoost
y_pred_xgb = best_xgb_model.predict(X_test)
print("\nClassification report pour XGBoost optimisÃ©:")
print(classification_report(y_test, y_pred_xgb))

# 16. Ãvaluation du modÃ¨le optimisÃ© LightGBM
y_pred_lgb = best_lgb_model.predict(X_test)
print("\nClassification report pour LightGBM optimisÃ©:")
print(classification_report(y_test, y_pred_lgb))

# 17. Courbes d'apprentissage pour XGBoost
train_sizes, train_scores, test_scores = learning_curve(best_xgb_model, X_train_resampled, y_train_resampled, cv=5)
plt.figure(figsize=(8,6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train Score')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Test Score')
plt.title("Courbes d'apprentissage pour XGBoost")
plt.xlabel("Taille de l'Ã©chantillon d'entraÃ®nement")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# 18. Courbes d'apprentissage pour LightGBM
train_sizes, train_scores, test_scores = learning_curve(best_lgb_model, X_train_resampled, y_train_resampled, cv=5)
plt.figure(figsize=(8,6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Train Score')
plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Test Score')
plt.title("Courbes d'apprentissage pour LightGBM")
plt.xlabel("Taille de l'Ã©chantillon d'entraÃ®nement")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
